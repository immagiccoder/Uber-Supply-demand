```python
# Import the pandas, Numpy, Matplotlib, Datetime and Seaborn packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

#Setting Seaborn Background style as White Grid

sns.set(style="whitegrid")
```

```python
# Read the csv file using 'read_csv'
uber=pd.read_csv('G:/Data science/Data_Sourcing/Uber Request Data.csv')
uber
```

```python
# Calculating Total Number of Demand By Counting All Request 
uber['Request id'].count()

```

```python
 # Extracting Request Date of uber booking from Request Timestamp
    
uber['Req_Date']=pd.DatetimeIndex(uber['Request timestamp']).date
uber
```

```python
# Extracting Drop Date of uber booking from Drop Timestamp
uber['drop_Date']=pd.DatetimeIndex(uber['Drop timestamp']).date
uber
```

```python
# Extracting Request Hour of uber booking from Request Timestamp

uber['Req_Hour']=pd.to_datetime(uber['Request timestamp'])
uber['Req_Hour']=uber['Req_Hour'].dt.hour
uber
```

```python
# Extracting Request Minute of uber booking from Request Timestamp
uber['Req_Minute']=pd.to_datetime(uber['Request timestamp'])
uber['Req_Minute']=uber['Req_Minute'].dt.minute
uber
```

```python
# Extracting Drop Hour of uber booking from Drop Timestamp
uber['Drop_Hour']=pd.to_datetime(uber['Drop timestamp'])
uber['Drop_Hour']=uber['Drop_Hour'].dt.hour
uber
```

```python
# Extracting Drop Minute of uber booking from Drop Timestamp
uber['Drop_Minute']=pd.to_datetime(uber['Drop timestamp'])
uber['Drop_Minute']=uber['Drop_Minute'].dt.minute
uber
```

```python
#Defining Time slots in Night, Early Morning,Morning,Daytime,Evening and Late Evening
uber['Time_Slot']=pd.cut(uber['Req_Hour'],[0,4,6,10,16,19,23],labels=['Night','Early Morning','Morning','Daytime','Evening','Late Evening'])
uber  
```

```python
#Plot Showing Status of uber cars at different Time Slots
sns.set(rc={'figure.figsize':(10,8)})
sns.countplot(x ="Time_Slot",hue="Status",data = uber)
```

```python
#Calculating total Cancelled Status for Morning Timeslot as Morning Timeslot have maximum cancellation
uber_morning_Cancelled=uber.loc[(uber['Status']=="Cancelled")&(uber['Time_Slot']=='Morning'),:]
uber_morning_Cancelled.Status.count()
```

```python
#Plot Showing uber Morning Timeslot have Maximum Cancelled Status
sns.set(rc={'figure.figsize':(10,8)})
sns.countplot(x="Time_Slot",hue="Status",data=uber_morning_Cancelled)
```

```python
#Calculating Evening Timeslot No Cars Available Status
uber_No_Cars_Available=uber.loc[(uber['Status']=="No Cars Available")&(uber['Time_Slot']=='Evening'),:]
uber_No_Cars_Available.Status.count()
```

```python
# Plot showing maximum number of No Cars Available for Evening Timeslot
sns.set(rc={'figure.figsize':(10,8)})
sns.countplot(x="Time_Slot",data=uber_No_Cars_Available)
```

```python
#Plot indicating Status of uber Cars for Airport to City and City to Airport 
sns.set(rc={'figure.figsize':(10,8)})
sns.countplot(x="Pickup point",hue="Status",data=uber)
```

```python
#Graph indicating Timeslots for uber Cars for Airport to City and City to Airport 
sns.set(rc={'figure.figsize':(10,8)})
sns.countplot(x="Pickup point",hue="Time_Slot",data=uber)
```

```python
#Calculating Total Number of No Car Availability for City
no_car_city=uber.loc[(uber['Status']=="No Cars Available")& (uber['Pickup point']=="City")]
uber['No_Cars_Available_City']=no_car_city.Status.count()
uber
```

```python
#Calculating Total Number of No Car Availability for Airport
no_car_airport=uber.loc[(uber['Status']=="No Cars Available")& (uber['Pickup point']=="Airport")]
uber['No_Cars_Available_Airport']=no_car_airport.Status.count()
uber
```

```python
# Plot Representing No Cars Availability Status for Airport to City in Different Timeslots (As at Airport maximum number of non availabilty of cars found)
sns.set(rc={'figure.figsize':(10,8)})
sns.countplot(x="No_Cars_Available_Airport",hue="Time_Slot",data=uber)
```

```python
# Calculating Total number of uber car Cancelled Status for Airport to City 
cancelled_city=uber.loc[(uber['Status']=="Cancelled")& (uber['Pickup point']=="Airport")]
uber['Cancelled_Airport']=cancelled_city.Status.count()
uber
```

```python
# Calculating Total number of uber car Cancelled Status for City to Airport
cancelled_city=uber.loc[(uber['Status']=="Cancelled")& (uber['Pickup point']=="City")]
uber['Cancelled_City']=cancelled_city.Status.count()
uber
```

```python
# Plot Representing Cancelled Status for City to Airport in Different Timeslots (As City to Airport have maximum number of Cancellation)
sns.set(rc={'figure.figsize':(10,8)})
sns.countplot(x="Cancelled_City",hue="Time_Slot",data=uber)
```

```python
#calculating Demand Supply and Gap

#Demand is Total Number of Request
uber['Total_Demand']=uber['Request id'].count()

#Total Supply is Total Trip Completed
Total_Supply=uber.loc[uber['Status']=="Trip Completed"].count()
uber['Total_Supply']=Total_Supply[0]
# Gap is subtraction of Demand and Supply
uber['gap']=uber['Total_Demand']-uber['Total_Supply']

uber
```

```python
# Plot for Gap 
sns.set(rc={'figure.figsize':(10,8)})
sns.countplot(x="gap",data=uber)
```

```python
# Calculating Supply for Different TimeSlot
# Early Morning Timeslot
Total_Supply_EMorning=uber.loc[(uber['Status']=="Trip Completed")&(uber['Time_Slot']=="Early Morning")]
Total_Supply_EMorning.Status.count()
```

```python
# Morning Timeslot
Total_Supply_Morning=uber.loc[(uber['Status']=="Trip Completed")&(uber['Time_Slot']=="Morning")]
Total_Supply_Morning.Status.count()
```

```python
# Daytime Timeslot
Total_Supply_Day=uber.loc[(uber['Status']=="Trip Completed")&(uber['Time_Slot']=="Daytime")]
Total_Supply_Day.Status.count()
```

```python
# Evening Timeslot
Total_Supply_Eve=uber.loc[(uber['Status']=="Trip Completed")&(uber['Time_Slot']=="Evening")]
Total_Supply_Eve.Status.count()
```

```python
# Late Evening Timeslot
Total_Supply_LEve=uber.loc[(uber['Status']=="Trip Completed")&(uber['Time_Slot']=="Late Evening")]
Total_Supply_LEve.Status.count()
```

```python
# Night Timeslot
Total_Supply_Night=uber.loc[(uber['Status']=="Trip Completed")&(uber['Time_Slot']=="Night")]
Total_Supply_Night.Status.count()
```

```python
# After Calculation we get Maximum Supply (618) at Morning Timeslot and Minimum Supply (174) at Night Timeslot

uber['Min_supply_Night']=Total_Supply_Night.Status.count()
uber['Max_Supply_Morning']=Total_Supply_Morning.Status.count()
uber
```

```python
#Plot for Maximum Demand Supply Gap for all Timeslot
uber['Max_Gap']=uber["Total_Demand"]-uber["Min_supply_Night"]
sns.set(rc={'figure.figsize':(10,8)})
sns.countplot(x="Max_Gap",hue="Time_Slot",data=uber)
```

```python
#Plot for Minimum Demand Supply Gap for all Timeslot
uber['Min_Gap']=uber["Total_Demand"]-uber["Max_Supply_Morning"]
sns.set(rc={'figure.figsize':(10,8)})
sns.countplot(x="Min_Gap",hue="Time_Slot",data=uber)
```
